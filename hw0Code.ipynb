```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as la
```

```python
mean = [0,0]
cov = [[1,0],[0,1]]

samples = np.random.multivariate_normal(mean, cov, 1000)

x1 = samples[:, 0]
x2 = samples[:, 1]

plt.figure(figsize=(6,6))
plt.scatter(x1, x2, s=10, alpha=0.5)
plt.xlabel('x1')
plt.ylabel('x2')

plt.title("Scatter plot")
plt.grid(True)
plt.show()
```
